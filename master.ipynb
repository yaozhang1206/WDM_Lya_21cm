```python
# play around at two different redshifts, and for close to line-of-sight
k_array = np.linspace(0.06, 1.4, 500)
z1 = 2.0 # redshift of observation
base_plot_z1 = np.zeros(len(k_array))
total_plot_z1 = np.zeros(len(k_array))
base_3keV_plot_z1 = np.zeros(len(k_array))
total_3keV_plot_z1 = np.zeros(len(k_array))

base_6keV_plot_z1 = np.zeros(len(k_array))
total_6keV_plot_z1 = np.zeros(len(k_array))

base_9keV_plot_z1 = np.zeros(len(k_array))
total_9keV_plot_z1 = np.zeros(len(k_array))

z5 = 4.0 # redshift of observation
base_plot_z5 = np.zeros(len(k_array))
total_plot_z5 = np.zeros(len(k_array))
base_3keV_plot_z5 = np.zeros(len(k_array))
total_3keV_plot_z5 = np.zeros(len(k_array))

base_6keV_plot_z5 = np.zeros(len(k_array))
total_6keV_plot_z5 = np.zeros(len(k_array))

base_9keV_plot_z5 = np.zeros(len(k_array))
total_9keV_plot_z5 = np.zeros(len(k_array))

for i in range(0,len(k_array)):
    base_plot_z1[i] = cdm.FluxP3D_lya_Mpc(z1, k_array[i], 0.1)
    total_plot_z1[i] = base_plot_z1[i] + cdm.FluxP3D_reio_Mpc(z1, k_array[i], 0.1)

    base_3keV_plot_z1[i] = wdm_3keV.FluxP3D_lya_Mpc(z1, k_array[i], 0.1)
    total_3keV_plot_z1[i] = base_3keV_plot_z1[i] + wdm_3keV.FluxP3D_reio_Mpc(z1, k_array[i], 0.1)

    base_6keV_plot_z1[i] = wdm_6keV.FluxP3D_lya_Mpc(z1, k_array[i], 0.1)
    total_6keV_plot_z1[i] = base_6keV_plot_z1[i] + wdm_6keV.FluxP3D_reio_Mpc(z1, k_array[i], 0.1)


    base_9keV_plot_z1[i] = wdm_9keV.FluxP3D_lya_Mpc(z1, k_array[i], 0.1)
    total_9keV_plot_z1[i] = base_9keV_plot_z1[i] + wdm_9keV.FluxP3D_reio_Mpc(z1, k_array[i], 0.1)

    
    base_plot_z5[i] = cdm.FluxP3D_lya_Mpc(z5, k_array[i], 0.1)
    total_plot_z5[i] = base_plot_z5[i] + cdm.FluxP3D_reio_Mpc(z5, k_array[i], 0.1)
    
    base_3keV_plot_z5[i] = wdm_3keV.FluxP3D_lya_Mpc(z5, k_array[i], 0.1)
    total_3keV_plot_z5[i] = base_3keV_plot_z5[i] + wdm_3keV.FluxP3D_reio_Mpc(z5, k_array[i], 0.1)

    base_6keV_plot_z5[i] = wdm_6keV.FluxP3D_lya_Mpc(z5, k_array[i], 0.1)
    total_6keV_plot_z5[i] = base_6keV_plot_z5[i] + wdm_6keV.FluxP3D_reio_Mpc(z5, k_array[i], 0.1)

    base_9keV_plot_z5[i] = wdm_9keV.FluxP3D_lya_Mpc(z5, k_array[i], 0.1)
    total_9keV_plot_z5[i] = base_9keV_plot_z5[i] + wdm_9keV.FluxP3D_reio_Mpc(z5, k_array[i], 0.1)

```

```python
# plotting CDM test 
fig, axs = plt.subplots(2, 4,figsize=(16,16), sharex=True)
axs[0,3].plot(k_array, base_plot_z1, '-', color='royalblue', label=r'lya, $z = 2$')
axs[0,3].plot(k_array, base_plot_z5, ':', color='magenta', label=r'lya, $z = 4$')
axs[0,3].plot(k_array, total_plot_z1, '--', color='purple', label=r'lya + reio, $z = 2$')
axs[0,3].plot(k_array, total_plot_z5, '-.', color='green', label=r'lya + reio, $z = 4$')

axs[0,3].grid(linestyle='dotted')
axs[0,3].set_title('CDM', fontsize=16)
# axs[0,3].set_yscale('log')
axs[0,3].set_xscale('log')
# axs[0,3].set_xlabel('k   [Mpc$^{-1}$]', fontsize=14)
axs[0,3].tick_params(axis = 'both', which = 'major', labelsize = 14)
# axs[0,3].set_ylabel(r'3D Flux power spectrum    [Mpc]', fontsize=16)
axs[0,3].legend(loc='best',fontsize=14)
axs[0,3].set_xlim(0.05, 1.5)
axs[0,3].set_ylim(0,900.)

# next wdm
axs[0,0].plot(k_array, base_3keV_plot_z1, '-', color='royalblue', label=r'lya, $z = 2$')
axs[0,0].plot(k_array, base_3keV_plot_z5, ':', color='magenta', label=r'lya, $z = 4$')
axs[0,0].plot(k_array, total_3keV_plot_z1, '--', color='purple', label=r'lya + reio, $z = 2$')
axs[0,0].plot(k_array, total_3keV_plot_z5, '-.', color='green', label=r'lya + reio, $z = 4$')

axs[0,0].grid(linestyle='dotted')
axs[0,0].set_title(r'm$_{\rm wdm} = 3 $ keV', fontsize=16)
# axs[0,0].set_yscale('log')
axs[0,0].set_xscale('log')
# axs[0,0].set_xlabel('k   [Mpc$^{-1}$]', fontsize=14)
axs[0,0].tick_params(axis = 'both', which = 'major', labelsize = 14)
axs[0,0].set_ylabel(r'3D Flux power spectrum    [Mpc]', fontsize=16)
# axs[0,0].legend(loc='best',fontsize=14)
axs[0,0].set_xlim(0.05, 1.5)
axs[0,0].set_ylim(0,900.)

# next wdm
axs[0,1].plot(k_array, base_6keV_plot_z1, '-', color='royalblue', label=r'lya, $z = 2$')
axs[0,1].plot(k_array, base_6keV_plot_z5, ':', color='magenta', label=r'lya, $z = 4$')
axs[0,1].plot(k_array, total_6keV_plot_z1, '--', color='purple', label=r'lya + reio, $z = 2$')
axs[0,1].plot(k_array, total_6keV_plot_z5, '-.', color='green', label=r'lya + reio, $z = 4$')

axs[0,1].grid(linestyle='dotted')
axs[0,1].set_title(r'm$_{\rm wdm} = 6 $ keV', fontsize=16)
# axs[0,1].set_yscale('log')
axs[0,1].set_xscale('log')
# axs[0,1].set_xlabel('k   [Mpc$^{-1}$]', fontsize=14)
axs[0,1].tick_params(axis = 'both', which = 'major', labelsize = 14)
# axs[0,1].set_ylabel(r'3D Flux power spectrum    [Mpc]', fontsize=16)
# axs[0,1].legend(loc='best',fontsize=14)
axs[0,1].set_xlim(0.05, 1.5)
axs[0,1].set_ylim(0,900.)

# next wdm
axs[0,2].plot(k_array, base_9keV_plot_z1, '-', color='royalblue', label=r'lya, $z = 2$')
axs[0,2].plot(k_array, base_9keV_plot_z5, ':', color='magenta', label=r'lya, $z = 4$')
axs[0,2].plot(k_array, total_9keV_plot_z1, '--', color='purple', label=r'lya + reio, $z = 2$')
axs[0,2].plot(k_array, total_9keV_plot_z5, '-.', color='green', label=r'lya + reio, $z = 4$')

axs[0,2].grid(linestyle='dotted')
axs[0,2].set_title(r'm$_{\rm wdm} = 9 $ keV', fontsize=16)
# axs[0,2].set_yscale('log')
axs[0,2].set_xscale('log')
# axs[0,2].set_xlabel('k   [Mpc$^{-1}$]', fontsize=14)
axs[0,2].tick_params(axis = 'both', which = 'major', labelsize = 14)
# axs[0,2].set_ylabel(r'3D Flux power spectrum    [Mpc]', fontsize=16)
# axs[0,2].legend(loc='best',fontsize=14)
axs[0,2].set_xlim(0.05, 1.5)
axs[0,2].set_ylim(0,900.)


axs[1,3].plot(k_array, 100.*np.abs(total_plot_z1 - base_plot_z1) / base_plot_z1,'--', color='purple', label=r'Planck-reio $z=2$')
axs[1,3].plot(k_array, 100.*np.abs(total_plot_z5 - base_plot_z5) / base_plot_z5,'-.', color='green', label=r'Planck-reio $z=4$')
axs[1,3].grid(linestyle='dotted')
axs[1,3].set_xscale('log')
axs[1,3].set_xlabel('k   [Mpc$^{-1}$]', fontsize=16)
axs[1,3].tick_params(axis = 'both', which = 'major', labelsize = 14)
# axs[1,3].set_ylabel(r'Frac. diff.    [%]', fontsize=16)
axs[1,3].set_xlim(0.05, 1.5)
axs[1,3].set_ylim(0., 105)

# next wdm
axs[1,0].plot(k_array, 100.*np.abs(total_3keV_plot_z1 - base_3keV_plot_z1) / base_3keV_plot_z1,'--', color='purple', label=r'reio $z=2$')
axs[1,0].plot(k_array, 100.*np.abs(total_3keV_plot_z5 - base_3keV_plot_z5) / base_3keV_plot_z5,'-.', color='green', label=r'reio $z=4$')
axs[1,0].grid(linestyle='dotted')
axs[1,0].set_xscale('log')
axs[1,0].set_xlabel('k   [Mpc$^{-1}$]', fontsize=16)
axs[1,0].tick_params(axis = 'both', which = 'major', labelsize = 14)
axs[1,0].set_ylabel(r'Frac. diff.    [%]', fontsize=16)
axs[1,0].set_xlim(0.05, 1.5)
axs[1,0].set_ylim(0., 105)

# next wdm
axs[1,1].plot(k_array, 100.*np.abs(total_6keV_plot_z1 - base_6keV_plot_z1) / base_6keV_plot_z1,'--', color='purple', label=r'reio $z=2$')
axs[1,1].plot(k_array, 100.*np.abs(total_6keV_plot_z5 - base_6keV_plot_z5) / base_6keV_plot_z5,'-.', color='green', label=r'reio $z=4$')
axs[1,1].grid(linestyle='dotted')
axs[1,1].set_xscale('log')
axs[1,1].set_xlabel('k   [Mpc$^{-1}$]', fontsize=16)
axs[1,1].tick_params(axis = 'both', which = 'major', labelsize = 14)
# axs[1,1].set_ylabel(r'Frac. diff.    [%]', fontsize=16)
axs[1,1].set_xlim(0.05, 1.5)
axs[1,1].set_ylim(0., 105)

# next wdm
axs[1,2].plot(k_array, 100.*np.abs(total_9keV_plot_z1 - base_9keV_plot_z1) / base_9keV_plot_z1,'--', color='purple', label=r'reio $z=2$')
axs[1,2].plot(k_array, 100.*np.abs(total_9keV_plot_z5 - base_9keV_plot_z5) / base_9keV_plot_z5,'-.', color='green', label=r'reio $z=4$')
axs[1,2].grid(linestyle='dotted')
axs[1,2].set_xscale('log')
axs[1,2].set_xlabel('k   [Mpc$^{-1}$]', fontsize=16)
axs[1,2].tick_params(axis = 'both', which = 'major', labelsize = 14)
# axs[1,2].set_ylabel(r'Frac. diff.    [%]', fontsize=16)
axs[1,2].set_xlim(0.05, 1.5)
axs[1,2].set_ylim(0., 105)


fig.tight_layout()
plt.savefig('3D_power_plot.pdf')
plt.show()
```

```python
# now we want a transparency plot
z_re = np.linspace(6.0, 12.0, 30)
psi_cdm_z1 = np.zeros(len(z_re))
psi_3keV_z1 = np.zeros(len(z_re))
psi_6keV_z1 = np.zeros(len(z_re))
psi_9keV_z1 = np.zeros(len(z_re))

psi_cdm_z5 = np.zeros(len(z_re))
psi_3keV_z5 = np.zeros(len(z_re))
psi_6keV_z5 = np.zeros(len(z_re))
psi_9keV_z5 = np.zeros(len(z_re))

for i in range(0, len(z_re)):
    psi_cdm_z1[i] = cdm.transparency(z_re[i], 2.0)
    psi_3keV_z1[i] = wdm_3keV.transparency(z_re[i], 2.0)
    psi_6keV_z1[i] = wdm_6keV.transparency(z_re[i], 2.0)
    psi_9keV_z1[i] = wdm_9keV.transparency(z_re[i], 2.0)

    
    psi_cdm_z5[i] = cdm.transparency(z_re[i], 4.0)
    psi_3keV_z5[i] = wdm_3keV.transparency(z_re[i], 4.0)
    psi_6keV_z5[i] = wdm_6keV.transparency(z_re[i], 4.0)
    psi_9keV_z5[i] = wdm_9keV.transparency(z_re[i], 4.0)
```

```python
# time to plot then

fig, axs = plt.subplots(1, 2,figsize=(12,6), sharey=True)
axs[0].plot(z_re,psi_cdm_z1,color='black',label=r'cdm')
axs[0].plot(z_re,psi_3keV_z1,color='purple',label=r'm$_{\rm wdm} = 3$ keV')
axs[0].plot(z_re,psi_6keV_z1,color='aqua',label=r'm$_{\rm wdm} = 6$ keV')
# axs[0].plot(z_re,psi_6keV_z1,color='aqua',label=r'm$_{\rm wdm} = 4$ keV')
axs[0].plot(z_re,psi_9keV_z1,color='chartreuse',label=r'm$_{\rm wdm} = 9$ keV')
axs[0].grid(linestyle='dotted')
axs[0].set_title(r'$z_{\rm obs} = 2.0$', fontsize=16)
axs[0].set_xlabel(r'$z_{\rm re}$',fontsize=16)
axs[0].set_ylabel(r'IGM Transparency    [%]',fontsize=16)
axs[0].legend(loc='best', fontsize=14)
axs[0].tick_params(axis = 'both', which = 'major', labelsize = 14)
# next
axs[1].plot(z_re,psi_cdm_z5,color='black',label=r'cdm')
axs[1].plot(z_re,psi_3keV_z5,color='purple',label=r'm$_{\rm wdm} = 3$ keV')
axs[1].plot(z_re,psi_6keV_z5,color='aqua',label=r'm$_{\rm wdm} = 6$ keV')
# axs[1].plot(z_re,psi_6keV_z5,color='aqua',label=r'm$_{\rm wdm} = 4$ keV')
axs[1].plot(z_re,psi_9keV_z5,color='chartreuse',label=r'm$_{\rm wdm} = 9$ keV')
axs[1].grid(linestyle='dotted')
axs[1].set_title(r'$z_{\rm obs} = 4.0$', fontsize=16)
axs[1].set_xlabel(r'$z_{\rm re}$',fontsize=16)
# axs[1].set_ylabel(r'Transparency    [%]',fontsize=16)
# axs[1].legend(loc='best', fontsize=14)
axs[1].tick_params(axis = 'both', which = 'major', labelsize = 14)
plt.savefig('transparency.pdf',bbox_inches="tight")
plt.show()
```
